```python
import numpy as np
import scipy as sp
import scgeom as sg
import scanpy as sc
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from scipy.spatial import distance_matrix
from scipy.stats import spearmanr, pearsonr, entropy

import matplotlib as mpl
plt.rcParams["font.family"] = "arial"
mpl.rcParams['font.size'] = 14
```

```python
adata = sc.read('./data/olsson.h5ad')
```

```python
sc.pp.neighbors(adata,metric = 'cosine',n_neighbors=10, use_rep='X')
sc.tl.umap(adata, random_state=547)
adata.obs['olsson_labels'] =  adata.obs['olsson_labels'].astype('str')
```

```python
perm_id_mutrans = sp.io.loadmat('./data/class_order.mat')['class_order'].flatten()
perm_class_mutrans = sp.io.loadmat('./data/perm_class.mat')['perm_class'].flatten()
```

```python
celltype_id_mutrans = -np.ones(adata.shape[0])
for i in range(adata.shape[0]):
    ii = np.where(perm_class_mutrans==i+1)[0]
    celltype_id_mutrans[i] = perm_id_mutrans[ii]
celltype_names_mutrans = ['HSPC-1','HSPC-2','Multi-lineage','Meg','Eryth','MDP','Mono','Mono-interm','Gran','Myeloid']
celltypes_mutrans = []
for i in range(adata.shape[0]):
    celltypes_mutrans.append(celltype_names_mutrans[int(celltype_id_mutrans[i])-1])
adata.obs['celltype_mutrans'] = pd.Series(list(celltypes_mutrans), dtype="category").values
```

```python
sc.pp.pca(adata)
X_pca = np.array( adata.obsm['X_pca'] )
X_umap = np.array( adata.obsm['X_umap'] )
```

```python
import plotly
colors = plotly.colors.qualitative.Plotly
celltypes = np.sort(list(set(adata.obs['celltype_mutrans'])))
celltype_vec = adata.obs['celltype_mutrans']
cnt = 0
for celltype in celltypes:
    idx = np.where(celltype_vec == celltype)[0]
    plt.scatter(X_umap[idx,0], X_umap[idx,1], c=colors[cnt], label=celltype)
    cnt += 1
plt.axis('equal')
plt.axis('off')
plt.grid(False)
plt.legend(loc=[1.01,0])
plt.tight_layout()
```

```python
from gudhi.wasserstein import wasserstein_distance
from gudhi import bottleneck_distance
def ph_feature_relative_homology_distance(
    adata,
    diagram_name = None,
    max_dim = 1,
    inf_value = 'replace_max',
    ph_distance = 'wasserstein',
    ph_wasserstein_p = 2,
):
    n = adata.shape[0]
    D = np.empty([n,max_dim+1])
    for i in range(n):
        max_filtration = adata.uns[diagram_name]['%s' % i]['max_filtration']
        for j in range(0, max_dim+1):
            dgm_i_rel = adata.uns[diagram_name]['%s' % i]['dgm_h%s' % j]
            dgm_i_base = adata.uns[diagram_name]['%s' % i]['dgm_h%s_base' % j]
            if inf_value == 'replace_max':
                dgm_i_rel[np.where(dgm_i_rel==np.inf)] = max_filtration
                dgm_i_base[np.where(dgm_i_base==np.inf)] = max_filtration
            if ph_distance == 'wasserstein':
                D[i,j] = wasserstein_distance(dgm_i_rel, dgm_i_base, internal_p=ph_wasserstein_p)
            elif ph_distance == 'bottleneck':
                D[i,j] = bottleneck_distance(dgm_i_rel, dgm_i_base)
            elif ph_distance == 'cumulative_length':
                D[i,j] = np.sum(dgm_i_rel[:,1]-dgm_i_rel[:,0]) - np.sum(dgm_i_base[:,1]-dgm_i_base[:,0])
    return D
```

```python
D = ph_feature_relative_homology_distance(adata, diagram_name='diagrams_cell_network_rlph', ph_distance='wasserstein',
    ph_wasserstein_p = 2)
```
